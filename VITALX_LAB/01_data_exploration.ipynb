```python
# ============================================
# PARAMETER RANGES VISUALIZATION
# ============================================

fig, axes = plt.subplots(2, 4, figsize = (16, 8))
fig.suptitle(
    'ASTROVITAL AI - VITALX CORE V1\nparameters Ranges From 12 Research Papers',
    fontsize = 14, fontweight = 'bold', color = 'white'
)

params_list = list(parameters.keys())
axes_flat = axes.flatten()

for idx, param in enumerate(params_list):
    ax = axes_flat[idx]
    p = parameters[param]

    # Background bar - full range
    ax.barh(0, p['max'] - p['min'],
            left = p['min'],
            color = '#FF4444', alpha = 0.4,
            height = 0.4, label = 'Full Range')
    
    # Normal Rnage
    ax.barh(0, p['normal_max'] - p['normal_min'],
            left=p['normal_min'],
            color='#00E676', alpha=0.8,
            height=0.4, label='Normal Range')
    
    ax.set_title(param.replace('_', ' '),
                 fontsize = 8, color = 'white')
    ax.set_xlabel(p['unit'], fontsize = 7, color = "gray")
    ax.set_yticks([])
    ax.legend(fontsize = 6)

plt.tight_layout()
plt.show()

print("✅ Graph 01 - Parameters Range showed Successfully.")
```

```python
# ============================================
# Visualization — Parameter Ranges
# ============================================

fig, ax = plt.subplots(figsize=(12, 6))

params = list(parameters.keys())
mins = [parameters[p]["min"] for p in params]
maxs = [parameters[p]["max"] for p in params]
ranges = [maxs[i] - mins[i] for i in range(len(params))]

bars = ax.barh(params, ranges, left=mins,
               color='cyan', alpha=0.7, edgecolor='white')

ax.set_xlabel('Parameter Value Range', color='white')
ax.set_title('ASTROVITAL AI — VITALX CORE V1\nPhysiological Parameter Ranges',
             color='gold', fontsize=13, fontweight='bold')
ax.tick_params(colors='white')

for i, (mn, mx) in enumerate(zip(mins, maxs)):
    ax.text(mn, i, f' {mn}', va='center', color='white', fontsize=8)
    ax.text(mx, i, f' {mx}', va='center', color='yellow', fontsize=8)

plt.tight_layout()
plt.show()
print("✅ Parameter ranges visualization complete!")
```
